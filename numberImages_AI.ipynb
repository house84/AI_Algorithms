```
mycolor = cv2.imread('Collard Green Stem (2).jpg')
resized = cv2.resize(mycolor, (256, 256), interpolation = cv2.INTER_AREA)
mygray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```
print(mygray.shape)
plt.imshow(mygray, cmap='gray')
plt.colorbar()
plt.show()
```
